```python
#Libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

First I will do feature engineering

```python
data_train = pd.read_csv('Train.csv')
data_test = pd.read_csv('Test.csv')
data = pd.concat([data_train, data_test], ignore_index=True)
data['income_above_limit'] = (data['income_above_limit'] == 'Above limit').astype(int)
```

Now I will go throught the data feature by feature.
I will remove columns which I think are not necessary to determine my output. One such feature is ID.

I will divide age into three categories using one hot encoding:

1.   Age(0 - 20)years (Children and teens)
2.   Age(20 - 50)years (Earning)
3.   Age(50 - 70)years (Retired)



```python
data['Age1'] = (data['age'] <= 20).astype(int)
data['Age2'] = ((data['age'] > 20) & (data['age'] <= 50)).astype(int)
data['Age3'] = (data['age'] > 50).astype(int)
```

```python
data['Gender'] = data['gender'].str.strip().replace({'Male': 1, 'Female': 0, 'Other': 0, 'male': 1})
```

```python
data['education'] = data['education'].str.strip()
```

```python
data['class_of_worker_missing'] = data['class'].isnull().astype(int)
```

```python
data['wage_label_encoded'] = np.where(data['wage_per_hour'] == 0, 0,  # Assign 0 for value 0
                                      np.where(data['wage_per_hour'] < 1000, 1, 2))
```

```python
data['TotalProfit'] = data['gains'] - data['losses'] + data['stocks_status']
```

```python
data['TotalProfit'] = (data['TotalProfit'] >=0).astype(int)
```

```python
# List of columns to label encode
categorical_columns = ['marital_status', 'race', 'employment_commitment','household_summary','tax_status','education','migration_code_change_in_msa',
                       'migration_prev_sunbelt','migration_code_move_within_reg','migration_code_change_in_reg','is_hispanic','residence_1_year_ago',
                       'under_18_family','country_of_birth_own','country_of_birth_mother','country_of_birth_father','citizenship'
                       ]

# Initialize a dictionary to store the label encoders for each column
label_encoders = {}

for column in categorical_columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le  # Store the label encoder for future use
```

```python
data = data.drop(['ID','class','old_residence_reg','stocks_status','losses',
                  'gains','occupation_code_main','industry_code_main','is_labor_union',
                  'wage_per_hour','unemployment_reason','education_institute','education',
                  'age','gender','household_stat','old_residence_reg','old_residence_state','veterans_admin_questionnaire'],axis = 1)
```

```python
import xgboost
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('income_above_limit', axis=1), data['income_above_limit'], test_size=0.33)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.info()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score

# Function to perform cross-validation
def cross_val(classifier, X_train, y_train, num_splits=10):
    scores = cross_val_score(classifier, X_train, y_train, cv=num_splits)
    print("Cross-Validation Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

classifier = XGBClassifier()

# Perform cross-validation
cross_val(classifier, X_train, y_train)

# Train the final model on the entire training set
final_model = classifier.fit(X_train, y_train)

# Evaluate the final model on the test set
test_accuracy = final_model.score(X_test, y_test)
print("Test Accuracy: %0.2f" % test_accuracy)
```

```python
# Predict labels for the test set
y_pred = final_model.predict(X_test)

# Create a DataFrame to store results
results = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

# Save results to CSV
results.to_csv('test_results.csv', index=False)
```
